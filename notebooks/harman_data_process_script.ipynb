```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the dataset
file_path = r'C:\Users\Harman\Documents\5000\captone\risk-train.txt'
data = pd.read_csv(file_path, sep='\t')  # Change sep if necessary (e.g., ',', ' ' for space)
data.dtypes
```

```python
# Convert some objects to int
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data['CLASS'] = label_encoder.fit_transform(data['CLASS'])
data['B_EMAIL'] = label_encoder.fit_transform(data['B_EMAIL'])
data['B_TELEFON'] = label_encoder.fit_transform(data['B_TELEFON'])
data['FLAG_LRIDENTISCH'] = label_encoder.fit_transform(data['FLAG_LRIDENTISCH'])
data['FLAG_NEWSLETTER'] = label_encoder.fit_transform(data['FLAG_NEWSLETTER'])
data['FAIL_RPLZ'] = label_encoder.fit_transform(data['FAIL_RPLZ'])
data['FAIL_RPLZORTMATCH'] = label_encoder.fit_transform(data['FAIL_RPLZORTMATCH'])
data['FAIL_RORT'] = label_encoder.fit_transform(data['FAIL_RORT'])
data['NEUKUNDE'] = label_encoder.fit_transform(data['NEUKUNDE'])
data['CHK_LADR'] = label_encoder.fit_transform(data['CHK_LADR'])
data['Z_LAST_NAME'] = label_encoder.fit_transform(data['Z_LAST_NAME'])
data['CHK_RADR'] = label_encoder.fit_transform(data['CHK_RADR'])
data['CHK_KTO'] = label_encoder.fit_transform(data['CHK_KTO'])
data['CHK_CARD'] = label_encoder.fit_transform(data['CHK_CARD'])
data['CHK_COOKIE'] = label_encoder.fit_transform(data['CHK_COOKIE'])
data['CHK_IP'] = label_encoder.fit_transform(data['CHK_IP'])
data['FAIL_LPLZ'] = label_encoder.fit_transform(data['FAIL_LPLZ'])
data['FAIL_LORT'] = label_encoder.fit_transform(data['FAIL_LORT'])
data['FAIL_LPLZORTMATCH'] = label_encoder.fit_transform(data['FAIL_LPLZORTMATCH'])

```

```python
#Replace missing values and convert into int
data['ANUMMER_02'] = pd.to_numeric(data['ANUMMER_02'], errors='coerce')
data['ANUMMER_02'] = data['ANUMMER_02'].fillna(0)
data['ANUMMER_02'] = data['ANUMMER_02'].astype(int)

data['ANUMMER_04'] = pd.to_numeric(data['ANUMMER_04'], errors='coerce')
data['ANUMMER_04'] = data['ANUMMER_04'].fillna(0)
data['ANUMMER_04'] = data['ANUMMER_04'].astype(int)

data['ANUMMER_03'] = pd.to_numeric(data['ANUMMER_03'], errors='coerce')
data['ANUMMER_03'] = data['ANUMMER_03'].fillna(0)
data['ANUMMER_03'] = data['ANUMMER_03'].astype(int)

data['ANUMMER_05'] = pd.to_numeric(data['ANUMMER_05'], errors='coerce')
data['ANUMMER_05'] = data['ANUMMER_05'].fillna(0)
data['ANUMMER_05'] = data['ANUMMER_05'].astype(int)

data['ANUMMER_06'] = pd.to_numeric(data['ANUMMER_06'], errors='coerce')
data['ANUMMER_06'] = data['ANUMMER_06'].fillna(0)
data['ANUMMER_06'] = data['ANUMMER_06'].astype(int)

data['ANUMMER_07'] = pd.to_numeric(data['ANUMMER_07'], errors='coerce')
data['ANUMMER_07'] = data['ANUMMER_07'].fillna(0)
data['ANUMMER_07'] = data['ANUMMER_07'].astype(int)

data['ANUMMER_08'] = pd.to_numeric(data['ANUMMER_08'], errors='coerce')
data['ANUMMER_08'] = data['ANUMMER_08'].fillna(0)
data['ANUMMER_08'] = data['ANUMMER_08'].astype(int)

data['ANUMMER_09'] = pd.to_numeric(data['ANUMMER_09'], errors='coerce')
data['ANUMMER_09'] = data['ANUMMER_09'].fillna(0)
data['ANUMMER_09'] = data['ANUMMER_09'].astype(int)

data['ANUMMER_10'] = pd.to_numeric(data['ANUMMER_10'], errors='coerce')
data['ANUMMER_10'] = data['ANUMMER_10'].fillna(0)
data['ANUMMER_10'] = data['ANUMMER_10'].astype(int)
```

```python
# Convert Z_CARD_VALID into int after replacing any missing values
data['Z_CARD_VALID'] = data['Z_CARD_VALID'].replace('?', np.nan)
mode_zcard = data['Z_CARD_VALID'].mode()[0] 
data['Z_CARD_VALID'] = data['Z_CARD_VALID'].fillna(mode_zcard)

data['Z_CARD_VALID'] = pd.to_numeric(data['Z_CARD_VALID'], errors='coerce')
data['Z_CARD_VALID'] = data['Z_CARD_VALID'].astype(int)
```

```python
data['VALUE_ORDER'] = data['VALUE_ORDER'].astype(int)
data['VALUE_ORDER_PRE'] = data['VALUE_ORDER_PRE'].astype(int)
```

```python
# Convert MAHN_AKT into int after replacing any missing values
data['MAHN_AKT'] = data['MAHN_AKT'].replace('?', np.nan)
data['MAHN_AKT'] = pd.to_numeric(data['MAHN_AKT'], errors='coerce')

mean_MAHN_AKT = data['MAHN_AKT'].mean()
data['MAHN_AKT'] = data['MAHN_AKT'].fillna(mean_MAHN_AKT)

data['MAHN_AKT'] = data['MAHN_AKT'].round().astype(int)
```

```python
# Convert MAHN_AKT into int after replacing any missing values
data['MAHN_HOECHST'] = data['MAHN_HOECHST'].replace('?', np.nan)
data['MAHN_HOECHST'] = data['MAHN_HOECHST'].astype(float)
mode_MAHN_HOECHST = data['MAHN_HOECHST'].mode()[0]

data['MAHN_HOECHST'] = data['MAHN_HOECHST'].fillna(mode_MAHN_HOECHST)
data['MAHN_HOECHST'] = data['MAHN_HOECHST'].astype(int)
```

```python
# Clean the TIME_ORDER column 
data['TIME_ORDER'] = pd.to_datetime(data["TIME_ORDER"] , format='%H:%M', errors='coerce')
data["TIME_ORDER"] = data["TIME_ORDER"] .dt.time.astype(str)

def time_to_seconds(time_str):
    if pd.isnull(time_str) or time_str in ['NaT','?']:
        return np.nan
    try:
        h, m, s = map(int, time_str.split(':'))
        return h * 3600 + m * 60 + s  
    except ValueError:
        return np.nan 
# Convert TIME_ORDER to total seconds
data['TIME_ORDER_SECONDS'] = data['TIME_ORDER'].apply(time_to_seconds)

# Calculate the mean
mean_seconds = data['TIME_ORDER_SECONDS'].mean(skipna=True)
# Fill NaN values in TIME_ORDER_SECONDS with the mean
data['TIME_ORDER_SECONDS'].fillna(mean_seconds, inplace=True)
# Convert TIME_ORDER_SECONDS to int
data['TIME_ORDER_SECONDS'] = data['TIME_ORDER_SECONDS'].astype(int)


# Convert total seconds back to datetime format
def seconds_to_time(seconds):
    if pd.isnull(seconds) or seconds < 0:
        return "00:00:00"  
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds = int(seconds % 60)
    return f"{hours:02}:{minutes:02}:{seconds:02}"

# Convert the mean back to datetime format
mean_time = seconds_to_time(mean_seconds)

# Fill null values with the mean time 
data['TIME_ORDER'] = data['TIME_ORDER'].fillna(mean_time)
# Make sure NaT is treated as NaN
data['TIME_ORDER'] = data['TIME_ORDER'].replace(pd.NaT, mean_time)

data['TIME_ORDER_SECONDS']


```

```python
# Make Z_CARD_ART int and replace missing values
card_type = {'Visa': 1,'Eurocard': 2,'debit_card': 3,'Amex': 4 }
data['Z_CARD_ART'] = data['Z_CARD_ART'].replace(card_type)

data['Z_CARD_ART'] = data['Z_CARD_ART'].replace('?', np.nan)
mode_z_card_art = data['Z_CARD_ART'].mode()[0]
data['Z_CARD_ART'] = data['Z_CARD_ART'].fillna(mode_z_card_art).astype(int)

print (f'Visa = 1\nEurocard = 2\nDebit Card = 3\nAmex = 4\n\n{data['Z_CARD_ART']}')
```

```python
# Make Z_METHODE int and replace missing values
payment_type = {'check': 1, 'credit_card' : 2 , 'debit_note': 3, 'debit_card' : 4}
data['Z_METHODE'] = data['Z_METHODE'].replace(payment_type)

data['Z_METHODE'] = data['Z_METHODE'].replace('?', np.nan)
mode_Z_METHODE = data['Z_METHODE'].mode()[0]
data['Z_METHODE'] = data['Z_METHODE'].fillna(mode_Z_METHODE).astype(int)

print (f'Check = 1\nCredit Card = 2\nDebit Note = 3\nDebit Card = 4\n\n{data['Z_METHODE']}')
```

```python
# Make WEEKDAY_ORDER int and replace missing values
days = {'Saturday': 1 , 'Wednesday': 2, 'Friday': 3, 'Tuesday' : 4, 'Thursday' : 5, 
        'Sunday' : 6, 'Monday' : 7}
data['WEEKDAY_ORDER'] = data['WEEKDAY_ORDER'].replace(days)

data['WEEKDAY_ORDER'] = data['WEEKDAY_ORDER'].replace('?', np.nan)
mode_WEEKDAY_ORDER = data['WEEKDAY_ORDER'].mode()[0]
data['WEEKDAY_ORDER'] = data['WEEKDAY_ORDER'].fillna(mode_WEEKDAY_ORDER).astype(int)
print(f'Saturday = 1\nWednesday = 2\nFriday = 3\nTuesday = 4\nThursday = 5\nSunday = 6\nMonday = 7\n\n{data['WEEKDAY_ORDER']}')
```

```python
# Clean B_BIRTHDATE - change to datetime format

data['B_BIRTHDATE'] = pd.to_datetime(data['B_BIRTHDATE'], format='%m/%d/%Y', errors='coerce')
data['B_BIRTHDATE'] = data['B_BIRTHDATE'].fillna(pd.Timestamp('1900-01-01'))
data['B_BIRTHDATE']
```

```python
# Clean DATE_LORDER - change to datetime format

data['DATE_LORDER'] = pd.to_datetime(data['DATE_LORDER'], format='%m/%d/%Y', errors='coerce')
data['DATE_LORDER'] = data['DATE_LORDER'].fillna(pd.Timestamp('1900-01-01'))
data['DATE_LORDER']
```

```python
# Find the correlation between all the variables except DATE_LORDER and B_BIRTHDATE
columns_to_check = ['CLASS', 'B_EMAIL', 'B_TELEFON', 'FLAG_LRIDENTISCH', 'FLAG_NEWSLETTER', 'Z_METHODE',
                    'Z_CARD_VALID', 'VALUE_ORDER', 'WEEKDAY_ORDER', 'AMOUNT_ORDER', 'ANUMMER_01',
                    'CHK_LADR', 'CHK_RADR', 'CHK_KTO', 'CHK_CARD', 'CHK_COOKIE', 'CHK_IP', 'FAIL_LPLZ',
                    'FAIL_LORT', 'FAIL_LPLZORTMATCH', 'FAIL_RPLZ', 'FAIL_RORT', 'FAIL_RPLZORTMATCH',
                    'SESSION_TIME', 'NEUKUNDE', 'AMOUNT_ORDER_PRE', 'VALUE_ORDER_PRE', 'TIME_ORDER_SECONDS']

int_columns = [col for col in columns_to_check if data[col].dtype in ['int32', 'int64']]
correlation_matrix = data[int_columns].corr()

filtered_corr_matrix = correlation_matrix[correlation_matrix >= 0]
print(filtered_corr_matrix)
```

```python
numeric_cols = data.select_dtypes(include=['int64', 'int32', 'float64'])
corr_with_class = numeric_cols.corr()['CLASS'].sort_values(ascending=False)

plt.figure(figsize=(10, 8))
sns.heatmap(corr_with_class.to_frame(), annot=True, cmap='coolwarm', 
            vmin=-1, vmax=1, cbar=True)
plt.title('Correlation of All Columns with CLASS')
plt.show()
```

```python
#spearman correlation 
numeric_cols = data.select_dtypes(include=['int64', 'int32', 'float64'])
spearman_corr = numeric_cols.corr(method='spearman')['CLASS'].sort_values(ascending=False)

plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr.to_frame(), annot=True, cmap='coolwarm', 
            vmin=-1, vmax=1, cbar=True)
plt.title('Correlation of All Columns with CLASS')
plt.show()
```
